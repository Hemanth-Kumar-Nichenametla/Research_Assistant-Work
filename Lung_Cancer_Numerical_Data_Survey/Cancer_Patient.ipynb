```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import pandas as pd

# Replace 'your_file.xlsx' with the path to your Excel file
file_path = 'cancer patient data sets.xlsx'

# Use the 'read_excel' function to load the Excel file into a DataFrame
df = pd.read_excel(file_path)

# Now, 'df' contains your data from the Excel file
```

```python
df
```

```python
#The column 'Level' refers to the disease severity
#Replacing strings (Low,Medium,High) in 'Level' with ints (1,2,3)
df['Level'] = df['Level'].replace('Low', 1)
df['Level'] = df['Level'].replace('Medium', 2) 
df['Level'] = df['Level'].replace('High', 3) 
df.head()
```

**In the above dataframe Patient Id is of no use while building the Machine Learning model. So, we can drop that column.**

```python
df = df.drop('Patient Id',axis=1)
df.head()
```

```python
# This code helps to check the Null values

df.isnull().sum()
```

```python
#Defining X(features) and y(response)
X=df[['Age', 'Gender', 'Air Pollution', 'Alcohol use', 'Dust Allergy',
       'OccuPational Hazards', 'Genetic Risk', 'chronic Lung Disease',
       'Balanced Diet', 'Obesity', 'Smoking', 'Passive Smoker']]
y=df['Level']

#scaling features with StandardScaler
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X)
scaled_X = scaler.transform(X)
scaled_X
```

```python
#splitting data into train and test 

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(scaled_X, y, test_size=0.30, random_state=16)
```

```python
#importing, training and evaluating KNN model
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

#creating a list of the K values that will be used to train the model
k = list(range(1,20,1))

#creating an empty list 'error_rate' that will append error rates of each model
error_rate = []

#for-loop to train the model with each value of K
for i in k:
    #setting K=i
    knn=KNeighborsClassifier(n_neighbors=i)
    #fitting model with train data
    knn.fit(X_train,y_train)
    #predicting values of test data
    y_pred = knn.predict(X_test)
    #appending into error_rate the error rates of the models
    error_rate.append(np.mean(y_pred != y_test))
    print(f'{i}:')
    print(accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    
    
#creating a plot that shows error rate vs k value.   
sns.lineplot(x = range(1,20),y=error_rate, marker="o")
```
